## Task 2: Visualizing the Target Distribution

```python
fig, ax = plt.subplots(figsize = (16, 6))
plt.subplot(1, 2, 1)
sns.distplot(Box_Office['revenue'], kde=False);
plt.title('Distribution of revenue');
```

```python
Box_Office.groupby('revenue')['revenue'].count().plot(kind='bar',figsize=(12,9),color='red')
```

## Task 3: Relationship between Film Revenue and Budget

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(Box_Office['budget'], Box_Office['revenue'])
plt.title('Revenue vs budget');
```

## Task 4: Distribution of Languages in Film

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='original_language', y='revenue', data=Box_Office.loc[Box_Office['original_language'].isin(Box_Office['original_language'].value_counts().head(10).index)]);
plt.title('Mean revenue per language');
```
